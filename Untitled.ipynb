```python
def weighted_bce(y_true, y_pred):
    LOSS_WEIGHT = 5
    loss_ones = LOSS_WEIGHT * y_true * tf.math.log(y_pred + 1e-7)
    loss_zeros = (1 - y_true) * tf.math.log(1 - y_pred + 1e-7)
    loss = - loss_ones - loss_zeros
    return tf.reduce_mean(loss)
```

```python
def getmodel(img_input):

    conv1=tf.keras.layers.Conv2D(64, 3, strides=2, activation="ReLU", padding="same")(img_input)

    conv2= tf.keras.layers.Conv2D(128,kernel_size=3,strides=2,activation="ReLU",padding="same")(conv1)

    conv3=tf.keras.layers.Conv2D(256,kernel_size=3,strides=2,activation="ReLU",padding="same")(conv2)

    conv4= tf.keras.layers.Conv2D(512,kernel_size=3,strides=2,activation="ReLU",padding="same")(conv3)

    conv5= tf.keras.layers.Conv2D(512,kernel_size=5,strides=1,activation="ReLU",padding="same")(conv4)

    conv6= tf.keras.layers.Conv2D(512,kernel_size=5,strides=1,activation="ReLU",padding="same")(conv5)

    l1 = tf.keras.layers.Concatenate()([conv5, conv6])

    convt1 =tf.keras.layers.Conv2DTranspose(512,kernel_size=1,strides=1,activation="ReLU")(l1)
    convt1=tf.keras.layers.BatchNormalization()(convt1)
    convt1 =tf.keras.layers.Conv2DTranspose(256,kernel_size=1,strides=1,activation="ReLU")(convt1)
    convt1=tf.keras.layers.BatchNormalization()(convt1)


    convt1 =tf.keras.layers.Conv2DTranspose(512,kernel_size=1,strides=1,activation="ReLU")(convt1)
    convt1=tf.keras.layers.BatchNormalization()(convt1)
    convt1 =tf.keras.layers.Conv2DTranspose(256,kernel_size=1,strides=1,activation="ReLU")(convt1)
    convt1=tf.keras.layers.BatchNormalization()(convt1)


    x=tf.keras.layers.Concatenate()([convt1,conv4])


    convt2 =tf.keras.layers.Conv2DTranspose(256,kernel_size=2,strides=2,activation="ReLU")(x)
    convt2=tf.keras.layers.BatchNormalization()(convt2)
    convt2 =tf.keras.layers.Conv2DTranspose(128,kernel_size=1,strides=1,activation="ReLU")(convt2)
    convt2=tf.keras.layers.BatchNormalization()(convt2)

    x=tf.keras.layers.concatenate(([convt2,conv3]))

    convt3 =tf.keras.layers.Conv2DTranspose(128,kernel_size=2,strides=2,activation="ReLU")(x)
    convt3=tf.keras.layers.BatchNormalization()(convt3)
    convt3 =tf.keras.layers.Conv2DTranspose(64,kernel_size=1,strides=1,activation="ReLU")(convt3)
    convt3=tf.keras.layers.BatchNormalization()(convt3)

    x=tf.keras.layers.concatenate(([convt3,conv2]))

    convt3 =tf.keras.layers.Conv2DTranspose(64,kernel_size=2,strides=2,activation="relu")(x)
    convt3=tf.keras.layers.BatchNormalization()(convt3)
    convt3 =tf.keras.layers.Conv2DTranspose(32,kernel_size=1,strides=1,activation="relu")(convt3)
    x=tf.keras.layers.BatchNormalization()(convt3)


    conv1=tf.keras.layers.Conv2D(32, 1, strides=1,activation="ReLU")(x)

    conv2= tf.keras.layers.Conv2D(32,kernel_size=1,strides=1,activation="ReLU")(x)

    conv3=tf.keras.layers.Conv2D(16,kernel_size=1,activation="ReLU")(conv2)

    x= tf.keras.layers.Conv2D(6,kernel_size=1,strides=1,activation="ReLU")(conv3)

 
    

    x=tf.keras.layers.Reshape((6,16384))(x)

    x =tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(6, recurrent_activation='sigmoid',return_sequences=True))(x)
    

    x=tf.keras.layers.Dense(20,activation="relu")(x)
    x=tf.keras.layers.Dense(20,activation="sigmoid")(x)

    #convt4 =tf.keras.layers.Conv2D(128,kernel_size=3,strides=1,activation="ReLU")(x)

    #conv1=tf.keras.layers.Conv2D(32, strides=1, activation="ReLU", padding="same",input_shape=(32,32,1))(x)

    #conv2= tf.keras.layers.Conv2D(32,kernel_size=3,strides=1,activation="ReLU",padding="same")(conv1)

    #conv3=tf.keras.layers.Conv2D(16,kernel_size=3,strides=1,activation="ReLU",padding="same",kernel_regularizer=l2(0.0005))(conv2)

    #conv4= tf.keras.layers.Conv2D(16,kernel_size=3,strides=1,activation="ReLU",padding="same")(conv3)
    return x

def t(row,colum,deep):
      img_input=keras.Input((row,colum,deep))
      x=getmodel(img_input)
      model = keras.Model(img_input,x, name='habal')
      opt = keras.optimizers.Adam(learning_rate=0.0002)
      model.compile(optimizer=opt,loss=tf.keras.losses.MeanAbsoluteError(),metrics=[tf.keras.metrics.Recall(),tf.keras.metrics.Precision()])
      model.build(X_train.shape)
      model.summary()
      return model
model=t(256,256,3)
```
